```python
import pandas as pd

tracks = pd.read_csv('./tracks_dataset.csv')
tracks.head()
```

```python
tracks = tracks.drop('Unnamed: 0', axis=1)
tracks.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

num_cols = [
    "popularity", "duration_ms", "danceability", "energy",
    "loudness", "speechiness", "acousticness", "instrumentalness",
    "liveness", "valence", "tempo"
]

scaler = MinMaxScaler()

tracks[num_cols] = scaler.fit_transform(tracks[num_cols])
tracks.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
tracks["artists"] = encoder.fit_transform(tracks["artists"])
tracks["album_name"] = encoder.fit_transform(tracks["album_name"])
tracks["track_genre"] = encoder.fit_transform(tracks["track_genre"])
tracks["explicit"] = tracks["explicit"].astype(int)
```

```python
# from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors

features = tracks.drop(columns=["track_id", "track_name"])
nn_model = NearestNeighbors(n_neighbors=10, metric="cosine", algorithm="auto")
nn_model.fit(features)

def get_similar_songs(track_index, n=10, artist_bias=0.1):
  track_features = features.iloc[track_index].values.reshape(1, -1)  # Convert to 2D array
  track_features_df = pd.DataFrame(track_features, columns=features.columns)  # Preserve column names

  distances, indices = nn_model.kneighbors(track_features_df, n_neighbors=n+1)

  similar_songs = tracks.iloc[indices[0][1:]].copy()
  original_artist = tracks.iloc[track_index]["artists"]
  
  similar_songs["adjusted_distance"] = [
    d - artist_bias if a == original_artist else d
    for d, a in zip(distances[0][1:], similar_songs["artists"])
  ]

  similar_songs = similar_songs.sort_values(by="adjusted_distance")

  return similar_songs.head(n)[["track_id", "track_name"]]

similar_songs = get_similar_songs(55002, n=10, artist_bias=0.9)
print(similar_songs)
```
